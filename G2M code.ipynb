```python
def plot_distribution(inp):
    plt.figure()
    ax=sns.distplot(inp)
    plt.axvline(np.mean(inp), color = "k", linestyle = "dashed", linewidth = 5)
    _, max_= plt.ylim() 
    plt.text(
        inp.mean() + inp.mean() / 10,
        max_ - max_ / 10,
        " Mean: {:.2f}".format(inp.mean()),
    )
    return plt.figure
```

# **Hypothesis**

```python
import datetime

import scipy.stats as stats
from sklearn.preprocessing import LabelEncoder
import copy
from pandas.api.types import CategoricalDtype
import statsmodels.api as sm
import statsmodels.formula.api as smf7
```

**1) The profit of yellow and pink cabs are almost equal**

```python
a = np.array(final[final.Company == "Yellow Cab"].Profit)  
b = np.array(final[final.Company == "Pink Cab"].Profit) 

_, p  = stats.ttest_ind(a,b, axis = 0) 

if p < 0.05: 
    print("Hypothesis is false with p value:",p)
else:
    print("Hypothesis is true with p value:",p)
```

**2)There is difference in profit between genders for yellow cabs**

```python
a= final[(final.Gender=='Female')&(final.Company=='Yellow Cab')].groupby('Transaction ID').Profit.mean()
b = final[(final.Gender=='Male')&(final.Company=='Yellow Cab')].groupby('Transaction ID').Profit.mean()

_, p = stats.ttest_ind(a.values,b.values,axis = 0)
if(p<0.05):
    print("Hypothesis is true with p value:",p)
else:
    print("Hypothesis is false with p value:",p)
    
```

**3)There is difference in profit between genders for pink cabs**

```python
a= final[(final.Gender=='Female')&(final.Company=='Pink Cab')].groupby('Transaction ID').Profit.mean()
b = final[(final.Gender=='Male')&(final.Company=='Pink Cab')].groupby('Transaction ID').Profit.mean()

_, p = stats.ttest_ind(a.values,b.values,axis = 0)
if(p<0.05):
    print("Hypothesis is true with p value:",p)
else:
    print("Hypothesis is false with p value:",p)
```

**4)Payment mode effects profits**

```python
a = np.array(final[final.Payment_Mode == "Cash"].Profit)  
b = np.array(final[final.Payment_Mode == "Card"].Profit) 

t, p  = stats.ttest_ind(a,b, axis = 0) 

if p < 0.05: 
    print("Payment Mode effects profit. Hypothesis is true.")
else:
    print("Payment Mode does not effect profit. Hypothesis is false.")
```

**5)Gender effects profit**

```python

a = np.array(final[final.Gender == "Male"].Profit)  
b = np.array(final[final.Gender == "Female"].Profit) 

t, p  = stats.ttest_ind(a,b, axis = 0)

if p < 0.05:  
    print("Gender effects profit. Hypothesis is true.")
else:
    print("Gender does not effect profit. Hypothesis is false.")
```

**6)Travel date has effect on preferred company.**

```python
c = pd.crosstab(final["Date of Travel"], final["Company"])  

_, p, __, ___ =  stats.chi2_contingency(c)

if p < 0.05: 
    print("Travel date has effect on company. Hypothesis is true.")
else:
    print("Travel date has no effect on company. Hypothesis is false.")
```

**7)Company and state distribution are not indipendent.**

```python

c = pd.crosstab(final["Company"], final["States"])  

_, p, __, ___ =  stats.chi2_contingency(c)

if p < 0.05:  # Setting our significance level at 5%
    print("Company and state distribution are not indipendent. Hypothesis is true.")
else:
    print("Company and state distribution are indipendent. Hypothesis is false.")
```
